# **2. Basic Operation Spark**

```python
# Membuat dataframe spark
data = [{"Nama" : 'Andre', "Umur" : '1'},
        {"Nama" : 'Badu', "Umur" : '2'},
        {"Nama" : 'Crish', "Umur" : '3'},
        {"Nama" : 'Ethan', "Umur" : '4'}]

df = spark.createDataFrame(data)
```

```python
# Menampilkan kontent dataframe dalam format tabular
df.show()
```

```python
# Collect --> Menampilkan konten struktur/metadata
df.collect()
```

```python
# Take --> Menampilkan konten struktur/metadata sebanyak sejumlah baris terbatas
df.take(3)
```

```python
# Count --> Menghitung jumlah total keseluruhan data
df.count()
```

```python
# Describe --> Mengembalikan nilai statistik dari dataframe
df.describe().show()
```

```python
# Summary --> 
df.summary().show()
```

```python
# Transformasi datafarame biasa menjadi RDD
df_rdd = df.rdd
```

```python
# Melihat jumlah cluster/partisi
df_rdd.getNumPartitions()
```

```python
# Melihat isi data yang sudah  di isi partisi
df_rdd.glom().collect()
```

```python

# Membuat data RDD menjadi DataFrame
df_biasa = df_rdd.toDF()
df_biasa.show()
```

# **a. Dataset**

```python
# Membaca Spark
# Header = True --> Digunakan untuk membuat row pertama data mejadi suatu nama fitur
# InferSchema =  True 
df_spark = spark.read.csv("karyawan.csv", header = True, inferSchema = True)

# Melihat isi dataset
df_spark.show()
```

```python
# Melihat tipe data jenis kolom
df_spark.dtypes
```

```python
# Melihat ifromasi /skema  dari data
df_spark.printSchema()
```

```python
# Memilih/Melihat informasi khusus pada suatu kolom 
df_spark.select("nama").show()
```

```python
# Memilih/Melihat informasi khusus pada suatu kolom 
df_spark.select("nama","gaji").show()
```

```python
# Memilih/Melihat informasi khusus pada suatu kolom 
df_spark.select("nama","gaji","pengalaman").show()
```

```python
# Transformasi data menambahkan kolom baru
df_spark = df_spark.withColumn("gaji_baru", df_spark["gaji"] + 1000000)
df_spark.show()
```

```python
# Transformasi hapus data
df_spark = df_spark.drop("gaji_baru")
df_spark.show()
```

```python
# Transformasi merubah nama/rename kolom
df_spark.withColumnRenamed("gaji", "salary").show()
```

```python
# Filtering data menggunakan filter
df_spark.filter("gaji <= 7000000").show()
```
